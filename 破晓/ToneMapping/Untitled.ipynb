```python
import numpy as np
import matplotlib.pyplot as plt

def Film(color,ACES_A, ACES_B, ACES_C, ACES_D, ACES_E):
    I = (ACES_A * color[:3] + ACES_B)
    F = (color[:3] * I)
    J = ACES_C * color[:3] + ACES_D
    G = (color[:3] * (J) + ACES_E)
    return np.clip(F / G, 0.0, 1.0)

X = np.arange(0, 5, 0.01)
custom_value = np.array([Film(np.array([x, x, x, 0]),1.36,0.0469,0.93,0.56,0.14) for x in X])
us_value = np.array([Film(np.array([x, x, x, 0]),2.51,0.03,2.43,0.59,0.14) for x in X])
plt.plot(X, custom_value[:,0], label='ABCDE = (1.36,0.0469,0.93,0.56,0.14)')
plt.plot(X, us_value[:,0], label='ABCDE = (2.51,0.03,2.43,0.59,0.14)')

plt.xlabel('Input luminance')
plt.ylabel('Output luminance')
plt.title('ACESFilm Functions')
plt.legend()
plt.grid(True)
plt.show()
```
